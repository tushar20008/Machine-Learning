```python
%matplotlib inline
import IPython.core.display         
# setup output image format (Chrome works best)
IPython.core.display.set_matplotlib_formats("svg")
import matplotlib.pyplot as plt
import matplotlib
from numpy import *
from sklearn import *
from scipy import stats
import IPython.utils.warn as warn
import csv
random.seed(100)
```

```python
def read_text_data(fname):
    txtdata = []
    classes = []
    topics  = []
    with open(fname, 'r') as csvfile:
        reader = csv.reader(csvfile, delimiter=',', quotechar='"')
        for row in reader:
            # get the text
            txtdata.append(row[0])
            # get the class (convert to integer)
            if len(row)>1:
                classes.append(row[1])
                topics.append(row[2])
    
    if (len(classes)>0) and (len(txtdata) != len(classes)):        
        warn.error("mismatched length!")
    
    return (txtdata, classes, topics)

def write_csv_kaggle_sub(fname, Y):
    # fname = file name
    # Y is a list/array with class entries
    
    # header
    tmp = [['Id', 'Prediction']]
    
    # add ID numbers for each Y
    for (i,y) in enumerate(Y):
        tmp2 = [(i+1), y]
        tmp.append(tmp2)
        
    # write CSV file
    f = open(fname, 'wb')
    writer = csv.writer(f)
    writer.writerows(tmp)
    f.close()
```

# Preprocessing - Clean the data

1) Word Stemming
    - Stem to data because tweets have similiar words with same root word. 
    - People use shotforms and miss spell the word, thus it increases the accuracy of the classifier.
    - The most common algorithm for stemming English is Porter's algorithm.
2) Lowercase
    - Convert all text to lowercase.
3) Remove unnecessary things
    - URL and username are removed using processTweet function.
    - Hashtag is removed for better representation of word.
    - Also, 2 or more characters that are reapated are removed.

```python
import re

def processTweet(tweet):
    # process the tweets

    #Convert www.* or https?://* to URL
    tweet = re.sub('((www\.[^\s]+)|(https?://[^\s]+))',' ',tweet)
    #Convert @username to AT_USER
    tweet = re.sub('@[^\s]+',' ',tweet)
    #Remove additional white spaces
    tweet = re.sub('[\s]+', ' ', tweet)
    #Replace #word with word
    tweet = re.sub(r'#([^\s]+)', r'\1', tweet)
    #trim
    tweet = tweet.strip('\'"')
    #Remove connectors
    tweet = re.sub('(\s+)(a|an|and|the)(\s+)', '\1\3', tweet)
    return tweet

def replaceTwoOrMore(s):
    #look for 2 or more repetitions of character and replace with the character itself
    pattern = re.compile(r"(.)\1{1,}", re.DOTALL)
    return pattern.sub(r"\1\1", s)

from nltk.stem import *
def clean_the_data(data):
    # Replacing punctuation chars with space
    dataTemp = data
    clean_txt= dataTemp
    ##[text1.replace(',',' ').replace('.',' ').replace('/',' ').replace('\\',' ').replace(';',' ').replace('\\',' ').replace('[',' ').replace(']',' ').replace('(',' ').replace(')',' ').replace('!',' ').replace('?',' ').replace('\'','') for text1 in dataTemp]
    
    for i, val in enumerate(clean_txt):
        processedTweet = processTweet(val)
        clean_txt[i] = processedTweet
    
    clean_txt = [replaceTwoOrMore(text1) for text1 in dataTemp]
    clean_txt = [text1.replace(',',' ').replace('.',' ').replace('/',' ').replace('\\',' ').replace(';',' ').replace('\\',' ').replace('[',' ').replace(']',' ').replace('(',' ').replace(')',' ').replace('!',' ').replace('?',' ').replace('\'','') for text1 in dataTemp]
    
#     # Ensuring that there is only one space between each word
    for i, val in enumerate(clean_txt):
        inp = val.split(' ')
        STEM = [PorterStemmer().stem(w.decode('utf-8')) for w in inp]
        tmp = ' '.join(STEM)
        clean_txt[i] = tmp

    clean_data = [text.lower() for text in clean_txt]

    return clean_data
```

# Observations:

- **The performace of Bernoulli classifier works best with Just count vectorizer and the performance drops with N_gram or TF-IDF. The reason for this is Naive Bayes uses the presence of a word (count vector) in a particular BoW model to classify and does not emphasis on weights for a particular word (TF-IDF). As Tf-IDF gives a weightage to determine the importance of a word, this particular weightage is not considered by Naive Bayes .**

The reason for this is MultinomialNB uses the presence of a word (count vector) in a particular BoW model to classify and emphasis on weights for a particular word (TF-IDF). As Tf-IDF gives a weightage to determine the importance of a word, this particular weightage is considered by MultinomialNB .**
